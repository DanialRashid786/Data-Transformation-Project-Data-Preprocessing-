#  Perform some aggregation on dataset

```python
df.head()
```

# Group by Department and calculate the average Annual Salary

```python
avg_salary_by_dept = df.groupby('Department')['Annual Salary'].mean()
print(avg_salary_by_dept)
```

# Group by Gender and calculate the total Bonus %:

```python
total_bonus_by_gender = df.groupby('Gender')['Bonus %'].sum()
print(total_bonus_by_gender)
```

# Group by Ethnicity and calculate the count of Employees:

```python
count_by_ethnicity = df.groupby('Ethnicity')['Employee ID'].count()
print(count_by_ethnicity)
```

# Group by Country and calculate the maximum Monthly Salary:

```python
max_salary_by_country = df.groupby('Country')['Monthly Salary'].max()
print(max_salary_by_country)
```

```python
# Group by Department and calculate mean salary and age
department_agg = df.groupby('Department').agg({
    'Annual Salary': 'mean',
    'Age': 'mean'
}).reset_index()
print(department_agg)
```

```python
# Create a pivot table to summarize average salary by Department and Gender
pivot_table = df.pivot_table(values='Annual Salary', index='Department', columns='Gender', aggfunc='mean')
print(pivot_table)
```

```python
# Group by Department and Business Unit, then calculate sum of salaries
multi_agg = df.groupby(['Department', 'Business Unit']).agg({
    'Annual Salary': 'sum'
}).reset_index()
print(multi_agg)
```

```python
# Define a custom aggregation function
def salary_range(series):
    return series.max() - series.min()

# Apply custom aggregation function
custom_agg = df.groupby('Department').agg({
    'Annual Salary': ['mean', 'sum', salary_range]
}).reset_index()
print(custom_agg)
```
